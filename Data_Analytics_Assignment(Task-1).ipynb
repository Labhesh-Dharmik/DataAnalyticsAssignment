```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Day': [f'Day {i}' for i in range(1, 11)], 
    'Page_Views': np.random.randint(1000, 5000, size=10),  
    'Average_Time_Spent': np.random.randint(30, 300, size=10),  
    'Bounce_Rate': np.random.uniform(0.2, 0.6, size=10) 
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Day'], df['Page_Views'], color='#091057')
plt.title('Page Views Over 10 Days')
plt.xlabel('Day')
plt.ylabel('Page Views')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
average_bounce_rate = df['Bounce_Rate'].mean()
engagement_rate = 1 - average_bounce_rate

```

```python
labels = ['Bounce Rate', 'Engagement']
sizes = [average_bounce_rate, engagement_rate]
colors = ['#091057', '#0D92F4']
explode = (0.1, 0)
```

```python
plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Bounce Rate vs Engagement')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Day'], df['Average_Time_Spent'], color='#0D92F4')
plt.title('Average Time Spent on the Page Over 10 Days')
plt.xlabel('Day')
plt.ylabel('Time Spent (Seconds)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("\nAverage Bounce Rate: {:.2f}%".format(average_bounce_rate * 100))
print("Engagement Rate: {:.2f}%".format(engagement_rate * 100))
```
